```python
import re

def remove_words(text: str, words_to_remove: set) -> str:
    filtered = ''
    for word in text.lower().strip('\n').split():
        word = re.sub('^[^a-z]*|[^a-z]*$', '', word)
        word = word.replace('’s', '').strip()
        if word != '' and word not in words_to_remove:
            filtered += word
            filtered += ' '
    return filtered;
```

```python
# load stop_words
def load_stop_words(fp: str) -> set:
    stop_words = set()
    with open(fp, 'r') as f:
        for line in f.readlines():
            stop_words.add(line.strip('\n'))
    return stop_words

fp = '/content/drive/MyDrive/news-data/stop_words.txt'
words_to_remove = load_stop_words(fp)
```

```python
# add english stop words
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

words_to_remove = words_to_remove.union(set(stopwords.words('english')))
len(words_to_remove)
```

```python
# define punctuations filter
import string
import re

def remove_punctuation(text: str):
    return text.translate(str.maketrans('', '', string.punctuation))

remove_punctuation('-.,/?nono!!#@\/"\'')
```

```python
import os 
import json
from tqdm import tqdm

drive_path = '/content/drive/MyDrive'
write_path = '/content/drive/MyDrive/news-split-data-processed'
read_path = '/content/drive/MyDrive/news-split-data'

with open('{}/train_text.json'.format(read_path), 'r') as f:
    raw_train_text = json.load(f)

with open('{}/test_text.json'.format(read_path), 'r') as f:
    raw_test_text = json.load(f)


proc_train_text = []
proc_test_text = []
for text in tqdm(raw_train_text):
    proc_train_text.append(remove_words(remove_punctuation(text), words_to_remove))

for text in tqdm(raw_test_text):
    proc_test_text.append(remove_words(remove_punctuation(text), words_to_remove))

with open('{}/train_text.json'.format(write_path), 'w') as f:
    json.dump(proc_train_text, f)

with open('{}/test_text.json'.format(write_path), 'w') as f:
    json.dump(proc_test_text, f)
```
